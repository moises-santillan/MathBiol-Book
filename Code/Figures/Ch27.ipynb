```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def Hill(x, n, K, kmax):
    return kmax * x**n / (x**n + K**n)
```

```python
fig, ax = plt.subplots()
P = np.linspace(0, 3, 200)
lhs = P

ax.plot(P, lhs, label='lhs', linewidth=2)

K = 2
kmax = 1
rhs = Hill(P, 3, K, kmax)
ax.plot(P, rhs, label=rf'rhs, $K={K}$', linewidth=2)

K = 1
kmax = 1
rhs = Hill(P, 3, K, kmax)
ax.plot(P, rhs, label=rf'rhs, $K={K}$', linewidth=2)

K = .5
kmax = 1
rhs = Hill(P, 3, K, kmax)
ax.plot(P, rhs, label=rf'rhs, $K={K}$', linewidth=2)

K = .2
kmax = 1
rhs = Hill(P, 3, K, kmax)
ax.plot(P, rhs, label=rf'rhs, $K={K}$', linewidth=2)

ax.set_ylim(0, 1.2)
ax.set_xlabel(r'$p$', fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=14)
ax.legend(fontsize=12)
fig.tight_layout()
fig.savefig('PDFs/fig-27.01.pdf')
fig.savefig('PNGs/fig-27.01.png')
```
