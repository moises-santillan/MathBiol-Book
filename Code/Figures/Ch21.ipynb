```python
import schemdraw
import schemdraw.elements as elm
import matplotlib
matplotlib.use('Agg')

d = schemdraw.Drawing()
R = elm.Resistor().right().label(r'$R$')
d.add(R)
C = elm.Capacitor().down().label(r'$C$')
d.add(C)
L = elm.Line().left()
d.add(L)
S = elm.SourceSin().up().label(r'$V_I(t)$')
d.add(S)
d.add(elm.LoopCurrent([R,C,L,S], pad = .6).label(r'$I(t)$'))
d.draw()
d.save('PDFs/fig-21.01.pdf')
d.save('PNGs/fig-21.01.png')
```
