```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
theta = np.linspace(np.pi, 10*np.pi, 200)
x = 0.2*(np.cos(theta) + 0.3 * theta)
y = 0.5 + 0.4*np.sin(theta)
fig, ax = plt.subplots()
ax.plot(x,y, color="k", linewidth=2)

ax.add_patch(plt.Rectangle((-.5, 0), .48, 2, facecolor='gray', edgecolor='black'))
ax.add_patch(plt.Rectangle((2.09, 0), 1, 1, facecolor='C1', edgecolor='black'))
plt.plot([-.5, 3.2], [-.02,-.02], "k", linewidth=3)

ax.arrow(2.4, 1.1, .5, 0, head_width=0.1, head_length=0.2, fc='k', ec='k', length_includes_head=True)
ax.arrow(2.9, -.15, -.5, 0, head_width=0.1, head_length=0.2, fc='k', ec='k', length_includes_head=True)

ax.text(1, 1.2, r'$K$', ha='center', va='center', fontsize=16)
ax.text(2.6, 0.5, r'$m$', ha='center', va='center', fontsize=16)
ax.text(2.6, 1.3, r'$v$', ha='center', va='center', fontsize=16)
ax.text(2.6, -.35, r'$F_f$', ha='center', va='center', fontsize=16)

ax.set_aspect('equal')
ax.set_axis_off()

plt.savefig("PDFs/fig-18.01.pdf")
plt.savefig("PNGs/fig-18.01.png")
```

```python
fig, ax = plt.subplots()

angle = np.pi/8
length = 2
# Draw the pivot and rod
ax.plot([0, (length-.2)*np.sin(angle)], [0, -(length-.2)*np.cos(angle)], 'k-', linewidth=2)
ax.add_artist(plt.Circle((0, 0), 0.1, fill=True, color='black'))

# Draw the mass
ax.add_artist(plt.Circle((length*np.sin(angle), -length*np.cos(angle)), 0.2, fill=True, color='C1'))

# Draw the angle theta
ax.plot([0, 0], [0, -0.75], 'k--')
ax.text(0.085, -0.5, r'$\theta$', color='k', ha='center', va='center', fontsize=14)

# Add labels
ax.text(length*np.sin(angle), -length*np.cos(angle), r'$m$', color='k', ha='center', va='center', fontsize=14)
ax.text(0.3, -1, r'$L$', color='k', ha='center', va='center', fontsize=14)

# Draw the force arrows
ax.arrow(.8, -2.1, 0, -0.5, color='C0', linewidth=2, head_width=0.1, head_length=0.2)
ax.text(.6, -2.3, r'$mg$', color='k', ha='center', va='center', fontsize=14)

ax.arrow(.72, -1.6, -0.5*np.sin(angle), .5*np.cos(angle), color='C0', linewidth=2, head_width=0.1, head_length=0.2)
ax.text(0.7, -1.3, r'$T$', color='k', ha='center', va='center', fontsize=14)

ax.set_xlim(-0.2, 1.2)
ax.set_aspect('equal')
ax.set_axis_off()

plt.savefig("PDFs/fig-18.02.pdf")
plt.savefig("PNGs/fig-18.02.png")
```
