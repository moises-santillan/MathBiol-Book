```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a new figure
fig, ax = plt.subplots(figsize=(4, 4))

# Set axis limits
ax.set_xlim([-1.2, 1.2])
ax.set_ylim([-1.2, 1.2])

# Draw the circle
ax.add_artist(plt.Circle((0, 0), 0.2, fill=True, color='orange'))

# Draw the arrows
ax.arrow(0, 0, 0, -1, head_width=0.1, head_length=0.2, fc='k', ec='k', length_includes_head=True)
ax.arrow(0, 0, 0, 1, head_width=0.1, head_length=0.2, fc='k', ec='k', length_includes_head=True)

# Add labels
ax.text(0, -1.1, r'$mg$', ha='center', va='top', fontsize=16)
ax.text(0, 1.1, r'$F_{r}$', ha='center', va='bottom', fontsize=16)

# Draw the point
ax.plot(0, 0, 'k.', markersize=10)

# Remove axes and ticks
ax.set_axis_off()

plt.savefig("PDFs/fig-02.01.pdf")
plt.savefig("PNGs/fig-02.01.png")
```

```python
# Create a new figure
fig, ax = plt.subplots()

# Draw the dashed line
ax.axhline(y=1.0, linestyle='dashed', color='k', linewidth=2)
ax.axvline(x=0, linestyle='dashed', color='k', linewidth=2)
ax.set_xlabel(r"$t$", fontsize=16)
ax.set_xticks([0])
ax.set_ylabel(r"$v(t)/v_T$", fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=14)

# Draw the solutions
vinits = [0.05, 0.25, 0.5, 1.5, 2.0, 2.5]
for vinit in vinits:
    t = np.linspace(0, 2, 500)
    v = 1.0 + (vinit - 1.0) * np.exp(-t / 0.5)
    ax.plot(t, v, color='C0', linewidth=2)
    
plt.savefig("PDFs/fig-02.02.pdf")
plt.savefig("PNGs/fig-02.02.png")
```

```python
fig, ax = plt.subplots()

# Draw the dashed line
ax.axhline(y=1, linestyle='dashed', color='k', linewidth=2)
ax.axvline(x=0, linestyle='dashed', color='k', linewidth=2)
ax.set_xlabel(r"$t$", fontsize=16)
ax.set_xticks([0])
ax.set_ylabel(r"$v(t)/v_T$", fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=14)

# Draw the exponential curves
tau_values = [0.25, 0.5, 1.0]
colors = ['C0', 'C1', 'C2']
legends = [r'$\tau/2$', r'$\tau$', r'$2\tau$']
for tau, color, legend in zip(tau_values, colors, legends):
    t = np.linspace(0, 2, 500)
    v = 1 + (2.0 - 1) * np.exp(-t / tau)
    ax.plot(t, v, color=color, linewidth=2, label=legend)
ax.legend(fontsize=14)

plt.savefig("PDFs/fig-02.03.pdf")
plt.savefig("PNGs/fig-02.03.png")
```

```python
# Create a new figure
fig, ax = plt.subplots(figsize=(4, 4))

# Draw the orange circle
circle = plt.Circle((0, 0), 0.2, color='orange', fill=True)
ax.add_artist(circle)

# Draw the force arrows
ax.arrow(-0.05, 0, 0, -1, head_width=0.1, head_length=0.2, fc='k', ec='k', length_includes_head=True)
ax.text(-.3, -1.1, r'$mg$', ha='left', va='center', fontsize=16)
ax.arrow(0.05, 0, 0, -1, head_width=0.1, head_length=0.2, fc='k', ec='k', length_includes_head=True)
ax.text(.3, -1.1, r'$f(t)$', ha='right', va='center', fontsize=16)
ax.arrow(0, 0, 0, 1, head_width=0.1, head_length=0.2, fc='k', ec='k', length_includes_head=True)
ax.text(0, 1.1, r'$F_{r}$', ha='center', va='center', fontsize=16)

# Draw the center point
ax.plot(0, 0, 'k.', markersize=10)

# Set axis limits and remove ticks
ax.set_xlim([-1.2, 1.2])
ax.set_ylim([-1.2, 1.2])
ax.set_axis_off()

plt.savefig("PDFs/fig-02.04.pdf")
plt.savefig("PNGs/fig-02.04.png")
```
