```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set the parameters
xmax = 2
ymax = 1.5

# Create a new figure
fig, ax = plt.subplots(figsize=(6, 6))

# Set axis limits
ax.set_xlim([-2, 2])
ax.set_ylim([-2, 2])

# Draw the parabola
y = np.linspace(-ymax, ymax, 500)
x = y ** 2
ax.plot(x, y, 'C1', linewidth=2, label=r"$\tau^2 = 4\Delta$")
plt.legend()
# Add the labels
ax.text(-0.5, 0, 'I', color='C0', fontsize=16)
ax.text(0.75, 1.5, 'II', color='C0', fontsize=16)
ax.text(1.5, 0.4, 'III', color='C0', fontsize=16)
ax.text(1.5, -0.4, 'IV', color='C0', fontsize=16)
ax.text(0.75, -1.5, 'V', color='C0', fontsize=16)

ax.set_xlabel(r"$\Delta$", fontsize=16)
ax.set_ylabel(r"$\tau$", fontsize=16)

ax.axvline(x=0, linestyle='--', color='C0', linewidth=2)
ax.plot([0, 2], [0,0], linestyle='--', color='C0', linewidth=2)

ax.set_xticks([0])
ax.set_yticks([0])
ax.tick_params(axis='both', which='major', labelsize=14)

plt.savefig("PDFs/fig-11.01.pdf")
plt.savefig("PNGs/fig-11.01.png")
```

```python
# Set the parameters
radius = 2
turns = 3
start_angle = 0
end_angle = turns * 2 * np.pi

# Create a new figure
fig, ax = plt.subplots(figsize=(6, 6))

# Draw the steady state
ax.scatter(0, 0, s=50, color='C1')

# Draw the trajectories
t = np.linspace(start_angle, end_angle, 500)
x1 = radius * np.exp(-0.2 * t) * np.cos(t)
y1 = radius * np.exp(-0.2 * t) * np.sin(t)
ax.plot(x1, y1, color='C0', linewidth=2)
ax.arrow(0, 1.46, .1, .02, head_width=0.1, head_length=0.2, fc='C0', ec='C0', length_includes_head=True)

x2 = radius * np.exp(-0.2 * t) * np.cos(t + np.pi)
y2 = radius * np.exp(-0.2 * t) * np.sin(t + np.pi)
ax.plot(x2, y2, color='C0', linewidth=2)
ax.arrow(0, -1.46, -.1, -.02, head_width=0.1, head_length=0.2, fc='C0', ec='C0', length_includes_head=True)

ax.set_xticks([])
ax.set_yticks([])
ax.set_title("(d)", fontsize=16)
```
