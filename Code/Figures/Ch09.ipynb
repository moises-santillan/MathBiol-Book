```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a new figure
fig, ax = plt.subplots()

# Draw the plots
ax.plot([0, 1], [1, 0], 'C0-', linewidth=2)
ax.plot([1, 2], [0, -1], 'C0--', linewidth=2)
ax.plot([0, 1], [0, 0], 'C1--', linewidth=2)
ax.plot([1, 2], [0, 0], 'C1-', linewidth=2)

ax.set_xlabel(r"$a$", fontsize=16)
ax.set_xticks([0, 1], [r'$0$', r'$r$'])
ax.set_ylabel(r"$N^*$", fontsize=16)
ax.set_yticks([0, 1], [r'$0$', r'$K$'])
ax.tick_params(axis='both', which='major', labelsize=14)
plt.tight_layout()
plt.savefig("PDFs/fig-09.01.pdf")
plt.savefig("PNGs/fig-09.01.png")
```

```python
# Create a new figure
fig, ax = plt.subplots()

# Draw the parabola
x = np.linspace(-0.25, 2.1, 100)
y = x * (1 - x/2) * (x/0.4 - 1)
ax.plot(x, y, 'C0-', linewidth=2)

ax.axhline(y=0, linestyle='--', color='k', linewidth=2)
ax.axvline(x=0, linestyle='dashed', color='k', linewidth=2)

ax.scatter([0, .4, 2], [0, 0, 0], s=50)

ax.set_xlabel(r"$N$", fontsize=16)
ax.set_xticks([0, .4, 2], [r'$N_1^*=0$', r'$N_2^*$', r'$N_3^*$'])
ax.set_ylabel(r"$f(N)$", fontsize=16)
ax.set_yticks([0])
ax.tick_params(axis='both', which='major', labelsize=14)
plt.tight_layout()
plt.savefig("PDFs/fig-09.02.pdf")
plt.savefig("PNGs/fig-09.02.png")
```
