```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a new figure
fig, ax = plt.subplots()

# Draw the parabola
x = np.linspace(-0.1, 1.1, 200)
y = x*(1-x)
ax.plot(x, y, 'C0', linewidth=2)

# Draw the points and arrows
ax.scatter([0, 1], [0, 0], s=80, facecolors='C0', edgecolors='C0')
ax.arrow(0.02, -0.01, .1, 0, head_width=0.01, head_length=0.025, fc='r', ec='r', length_includes_head=True)
ax.arrow(-0.02, -0.01, -.1, 0, head_width=0.01, head_length=0.025, fc='r', ec='r', length_includes_head=True)
ax.arrow(1.12, -0.01, -.1, 0, head_width=0.01, head_length=0.025, fc='r', ec='r', length_includes_head=True)
ax.arrow(0.88, -0.01, .1, 0, head_width=0.01, head_length=0.025, fc='r', ec='r', length_includes_head=True)


ax.axhline(y=0, linestyle='dashed', color='k', linewidth=2)
ax.set_xlabel(r"$x$", fontsize=16)
ax.set_ylabel(r"$f(x)$", fontsize=16)
ax.set_xticks([0, 1], [r"$x_1$", r"$x_2$"])
ax.set_yticks([0])
ax.tick_params(axis='both', which='major', labelsize=14)

plt.savefig("PDFs/fig-06.01.pdf")
plt.savefig("PNGs/fig-06.01.png")
```
