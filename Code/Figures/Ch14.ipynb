```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Set the figure size
fig, ax = plt.subplots(figsize=(6, 2))

# Define the box dimensions and positions
box_width = 1
box_height = 1
x_spacing = 2

# Draw the boxes
ax.add_patch(plt.Rectangle((0, 0), box_width, box_height, facecolor='C0', edgecolor='black'))
ax.text(box_width/2, box_height/2, r'$S$', ha='center', va='center', fontsize=16)

ax.add_patch(plt.Rectangle((x_spacing, 0), box_width, box_height, facecolor='C1', edgecolor='black'))
ax.text(x_spacing + box_width/2, box_height/2, r'$I$', ha='center', va='center', fontsize=16)

ax.add_patch(plt.Rectangle((2*x_spacing, 0), box_width, box_height, facecolor='C2', edgecolor='black'))
ax.text(2*x_spacing + box_width/2, box_height/2, r'$R$', ha='center', va='center', fontsize=16)

# Draw the arrows
ax.arrow(-x_spacing/2, box_width/2, x_spacing/2, 0, head_width=0.1, head_length=0.3, fc='k', ec='k', length_includes_head=True)
ax.text(-x_spacing/4, box_height/2+0.3, r'$\alpha$', ha='center', va='center', fontsize=14)

ax.arrow(x_spacing/2, box_height/2, x_spacing/2, 0, head_width=0.1, head_length=0.3, fc='k', ec='k', length_includes_head=True)
ax.text(x_spacing*.75, box_height/2+0.3, r'$\beta SI$', ha='center', va='center', fontsize=14)

ax.arrow(x_spacing*3/2, box_height/2, x_spacing/2, 0, head_width=0.1, head_length=0.3, fc='k', ec='k', length_includes_head=True)
ax.text(x_spacing*1.75, box_height/2+0.3, r'$\gamma I$', ha='center', va='center', fontsize=14)

ax.arrow(x_spacing/4, 0, 0, -x_spacing/4, head_width=0.1, head_length=0.3, fc='k', ec='k')
ax.text(x_spacing/4+.1, -x_spacing/8, r'$\mu S$', ha='left', va='center', fontsize=14)

ax.arrow(x_spacing*5/4, 0, 0, -x_spacing/4, head_width=0.1, head_length=0.3, fc='k', ec='k')
ax.text(x_spacing*5/4+.1, -x_spacing/8, r'$(\mu+\delta) I$', ha='left', va='center', fontsize=14)

ax.arrow(x_spacing*9/4, 0, 0, -x_spacing/4, head_width=0.1, head_length=0.3, fc='k', ec='k')
ax.text(x_spacing*9/4+.1, -x_spacing/8, r'$\mu R$', ha='left', va='center', fontsize=14)

# Remove axes and ticks
ax.set_axis_off()

plt.savefig("PDFs/fig-14.01.pdf")
plt.savefig("PNGs/fig-14.01.png")
```

```python
# Set the figure size
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Plot the first figure
ax1.set_xlim([0, 3])
ax1.set_ylim([0, 3])
ax1.set_xlabel(r'$R_0$', fontsize=16)
ax1.set_ylabel(r'$s^*$', fontsize=16)
ax1.tick_params(axis='both', which='major', labelsize=14)

ax1.plot([0, 1], [1, 1], color='C0', linewidth=2)
ax1.plot([1, 3], [1, 1], color='C0', linestyle='--', linewidth=2)
ax1.plot(np.linspace(1/3, 1, 100), [1/x for x in np.linspace(1/3, 1, 100)], color='C1', linestyle='--', linewidth=2)
ax1.plot(np.linspace(1, 3, 100), [1/x for x in np.linspace(1, 3, 100)], color='C1', linewidth=2)
ax1.set_aspect('equal')

# Plot the second figure
ax2.set_xlim([0, 3])
ax2.set_ylim([-1.5, 1])
ax2.set_xlabel(r'$R_0$', fontsize=16)
ax2.set_ylabel(r'$i^*$', fontsize=16)
ax2.tick_params(axis='both', which='major', labelsize=14)

ax2.plot([0, 1], [0, 0], color='C0', linewidth=2)
ax2.plot([1, 3], [0, 0], color='C0', linestyle='--', linewidth=2)
ax2.plot(np.linspace(2/5, 1, 100), [1 - 1/x for x in np.linspace(2/5, 1, 100)], color='C1', linestyle='--', linewidth=2)
ax2.plot(np.linspace(1, 3, 100), [1 - 1/x for x in np.linspace(1, 3, 100)], color='C1', linewidth=2)
ax2.set_aspect('equal')

plt.tight_layout()

plt.savefig("PDFs/fig-14.02.pdf")
plt.savefig("PNGs/fig-14.02.png")
```
