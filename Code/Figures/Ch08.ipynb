```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a new figure
fig, ax = plt.subplots()

# Draw the parabola
x = np.linspace(-0.1, 1.1, 200)
y = x*(1-x)
ax.plot(x, y, 'C0', linewidth=2)

# Draw the points and arrows
ax.scatter(0, 0, s=80, facecolors='none', edgecolors='C0')
ax.scatter(1, 0, s=80, facecolors='C0', edgecolors='C0')

ax.axhline(y=0, linestyle='dashed', color='k', linewidth=2)
ax.axvline(x=0, linestyle='dashed', color='k', linewidth=2)
ax.set_xlabel(r"$N$", fontsize=16)
ax.set_xticks([0, 1], [r'$0$', r'$K$'])
ax.set_ylabel(r"$f(N)$", fontsize=16)
ax.set_yticks([0])
ax.tick_params(axis='both', which='major', labelsize=14)

plt.savefig("PDFs/fig-08.01.pdf")
plt.savefig("PNGs/fig-08.01.png")
```

```python
# Create a new figure
fig, ax = plt.subplots(figsize=(8, 6))

# Draw the logistic functions
x = np.linspace(0, 5, 500)
ax.plot(x, 1 / (1 + (1 / 0.01 - 1) * np.exp(-2 * x)), 'C0', linewidth=2)
ax.plot(x, 1 / (1 + (1 / 0.1 - 1) * np.exp(-2 * x)), 'C0', linewidth=2)
ax.plot(x, 1 / (1 + (1 / 0.3 - 1) * np.exp(-2 * x)), 'C0', linewidth=2)
ax.plot(x, 1 / (1 + (1 / 0.6 - 1) * np.exp(-2 * x)), 'C0', linewidth=2)
ax.plot(x, 1 / (1 + (1 / 0.8 - 1) * np.exp(-2 * x)), 'C0', linewidth=2)
ax.plot(x, 1 / (1 + (1 / 1.5 - 1) * np.exp(-2 * x)), 'C0', linewidth=2)
ax.plot(x, 1 / (1 + (1 / 2 - 1) * np.exp(-2 * x)), 'C0', linewidth=2)
ax.plot(x, 1 / (1 + (1 / 2.9 - 1) * np.exp(-2 * x)), 'C0', linewidth=2)

# Draw the dashed lines
ax.axhline(y=1, linestyle='--', color='k', linewidth=2)
ax.axhline(y=0.5, linestyle='--', color='k', linewidth=2)
ax.axvline(x=0, linestyle='dashed', color='k', linewidth=2)
ax.set_xlabel(r"$t$", fontsize=16)
ax.set_xticks([0])
ax.set_ylabel(r"$f(N)$", fontsize=16)
ax.set_yticks([0.5, 1], [r"$K/2$", r"$K$"])
ax.tick_params(axis='both', which='major', labelsize=14)

plt.savefig("PDFs/fig-08.02.pdf")
plt.savefig("PNGs/fig-08.02.png")
```
