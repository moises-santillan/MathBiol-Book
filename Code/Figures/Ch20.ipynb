```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create a new figure
fig, ax = plt.subplots(figsize=(6, 6))

# Draw the axes
ax.axvline(x=0, linestyle='--', color='k', linewidth=2)
ax.axhline(y=0, linestyle='--', color='k', linewidth=2)

# Define the range for x
x = np.linspace(-0.25, 3.1, 400)

# Nullcline I = 0
y1 = x * (1 - x) * (x - 3)
ax.plot(x, y1, color='C1', linewidth=2, label=r'$I = 0$')

# Nullcline I = I_1 > 0
y2 = x * (1 - x) * (x - 3) + 1
ax.plot(x, y2, color='C2', linewidth=2, label=r'$I = 1$')

# Nullcline I = I_2 > I_1
y3 = x * (1 - x) * (x - 3) + 2
ax.plot(x, y3, color='C3', linewidth=2, label=r'$I = 2$')

# R nullcline
y4 = 2.2 * x
ax.plot(x[x <= 2], y4[x <= 2], color='C0', linewidth=2, label=r'$R$ nullcline')

# Add legend
ax.legend(loc='upper left', fontsize=14)

# Show the plot
ax.set_xlabel(r'$V$', fontsize=16)
ax.set_ylabel(r'$R$', fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=14)

fig.savefig("PDFs/fig-20.02.pdf")
fig.savefig("PNGs/fig-20.02.png")
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

# Define the domain for x and y
x = np.linspace(-2.5, 5.5, 30)
y = np.linspace(-40, 40, 30)

# Create a meshgrid for the vector field
X, Y = np.meshgrid(x, y)

# Define the vector field based on the given dynamics
U = X*(1-X)*(X-3)-Y+2
V = (2.2*X-Y)

# Create the quiver plot
ax[0].quiver(X, Y, U, V, color='olive', scale=1e3)


# Plot the nullcline
nullcline_x = np.linspace(-2.2, 5, 400)
nullcline_y = nullcline_x * (1 - nullcline_x) * (nullcline_x - 3) + 2
ax[0].plot(nullcline_x, nullcline_y, color='C0', lw=2, label='V-Nullcline')
ax[0].plot(nullcline_x, 2.2*nullcline_x, color='C1', lw=2, label='R-Nullcline')

# Set the limits and labels
ax[0].set_title(r'$\epsilon=1$', fontsize=16)
ax[0].set_xlabel(r'$V$', fontsize=16)
ax[0].set_ylabel(r'$R$', fontsize=16)
ax[0].grid(False)

# Define the domain for x and y
x = np.linspace(-2.5, 5.5, 30)
y = np.linspace(-40, 40, 30)

# Create a meshgrid for the vector field
X, Y = np.meshgrid(x, y)

# Define the vector field based on the given dynamics
U = X*(1-X)*(X-3)-Y+2
V = (2.2*X-Y)/100

# Create the quiver plot
ax[1].quiver(X, Y, U, V, color='olive', scale=2e3)


# Plot the nullcline
nullcline_x = np.linspace(-2.2, 5, 400)
nullcline_y = nullcline_x * (1 - nullcline_x) * (nullcline_x - 3) + 2
ax[1].plot(nullcline_x, nullcline_y, color='C0', lw=2, label='V-Nullcline')
ax[1].plot(nullcline_x, 2.2*nullcline_x, color='C1', lw=2, label='R-Nullcline')

# Set the limits and labels
ax[1].set_title(r'$\epsilon=100$', fontsize=16)
ax[1].set_xlabel('$V$', fontsize=16)
#ax[1].set_ylabel('$R$', fontsize=16)
ax[1].grid(False)
fig.tight_layout()
fig.savefig("PDFs/fig-20.03.pdf")
fig.savefig("PNGs/fig-20.03.png")
```

```python
# Set up the figure
fig, ax = plt.subplots(figsize=(6, 6))

# Define the domain for x and y
x = np.linspace(-0.75, 3.5, 30)
y = np.linspace(-2.5, 8, 30)

# Create a meshgrid for the vector field
X, Y = np.meshgrid(x, y)

# Define the vector field based on the given dynamics
U = X*(1-X)*(X-3)-Y+2
V = (2.2*X-Y)/100

# Create the quiver plot
ax.quiver(X, Y, U, V, color='olive', scale=2e2)


# Plot the nullcline
nullcline_x = np.linspace(-0.75, 3.5, 400)
nullcline_y = nullcline_x * (1 - nullcline_x) * (nullcline_x - 3) + 2
ax.plot(nullcline_x, nullcline_y, color='C0', lw=2, label='V-Nullcline')
ax.plot(nullcline_x, 2.2*nullcline_x, color='C1', lw=2, label='R-Nullcline')

# Draw blue arrows and lines
ax.arrow(0.45, 1.35, 2.65, 0, lw=2, head_width=0.1, head_length=0.2, color='C2', length_includes_head=True)
x_aux = np.linspace(3.1, 2.2, 100)
ax.plot(x_aux, x_aux * (1 - x_aux) * (x_aux - 3) + 2, color='C2', linewidth=2)

ax.arrow(2.2, 4.12, -2.62, 0, lw=2, head_width=0.1, head_length=0.2, color='C2', length_includes_head=True)
x_aux = np.linspace(-0.435, 0.45, 100)
ax.plot(x_aux, x_aux * (1 - x_aux) * (x_aux - 3) + 2, color='C2', linewidth=2)


# Set the limits and labels
ax.set_title(r'$\epsilon=100$', fontsize=16)
ax.set_xlabel('$V$', fontsize=16)
ax.set_ylabel('$R$', fontsize=16)
ax.grid(False)

# Show the plot
fig.legend(fontsize=14)
ax.tick_params(axis='both', which='major', labelsize=14)

fig.savefig("PDFs/fig-20.04.pdf")
fig.savefig("PNGs/fig-20.04.png")
```
